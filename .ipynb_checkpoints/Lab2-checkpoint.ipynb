```python
# Importăm bibliotecile necesare
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score

# Definim calea către fișierul CSV
file_path = r'C:\Users\Admin\obesity.csv'

# Încărcăm setul de date din fișierul CSV
obes = pd.read_csv(file_path)

# Replace 'male' with 1 and 'female' with 0
obes['Gender'] = obes['Gender'].replace({'Male': 1, 'Female': 0})
# Replace obesity categories with numerical values
obes['ObesityCategory'] = obes['ObesityCategory'].replace({
    'Underweight': 0,
    'Normal weight': 1,
    'Overweight': 2,
    'Obese': 3
})
```

```python
# Selecția caracteristicelor (atributelor) și etichetelor (claselor)
X = obes[['Age', 'Gender', 'Height', 'Weight', 'BMI', 'PhysicalActivityLevel']]
y = obes['ObesityCategory']

# Impartim setul de date in setul de antrenare si setul de testare
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)

# 1. Clasificatorul Naive Bayes
naive_bayes = GaussianNB()
naive_bayes.fit(X_train, y_train)

y_pred_prob_nb = naive_bayes.predict_proba(X_test) #calculează estimările de probabilitate pentru 
#fiecare clasă pentru fiecare eșantion din setul de testare.
y_pred_nb = naive_bayes.predict(X_test)

accuracy_nb = accuracy_score(y_test, y_pred_nb)
print("Accuracy Naive Bayes:", accuracy_nb)
```

- Având în vedere setul de date X_train și y_train, algoritmul Naive Bayes calculează probabilitățile pe baza frecvenței caracteristicilor din fiecare clasă.
- Pentru fiecare caracteristică, acesta calculează media și varianța pentru fiecare clasă. Aceasta înseamnă că, pentru caracteristicile continue, cum ar fi vârsta, înălțimea, greutatea etc., algoritmul calculează media și varianța acestor caracteristici pentru fiecare clasă în parte.
- De exemplu, dacă aveți patru clase în variabila dvs. țintă "ObesityCategory" (să spunem clasa 0, clasa 1, clasa 2 și clasa 3), algoritmul va calcula media și varianța "Age" pentru fiecare dintre aceste clase și, în mod similar, pentru alte caracteristici precum "Height", "Weight", "BMI" și "PhysicalActivityLevel".
- Aceste statistici sunt învățate din datele de formare și sunt utilizate în timpul fazei de testare pentru a calcula probabilitatea de a observa fiecare caracteristică dată de clasă.

- După ce modelul este antrenat, acesta este utilizat pentru a prezice etichetele de clasă pentru setul de date de testare (X_test).
Pentru fiecare eșantion din setul de testare, modelul calculează probabilitatea fiecărei clase, având în vedere caracteristicile observate, folosind statisticile învățate (media și varianța).
Acesta face acest lucru prin aplicarea teoremei lui Bayes, care presupune înmulțirea probabilității anterioare a fiecărei clase cu probabilitatea de a observa caracteristicile date de clasa respectivă.
În cele din urmă, selectează clasa cu cea mai mare probabilitate ca etichetă de clasă prezisă pentru fiecare eșantion.

```python
import numpy as np

# Sum up the probabilities for each class across all samples
class_probabilities_sum = np.sum(y_pred_prob_nb, axis=0)

# Normalize the sum to obtain the overall probability distribution across classes
overall_class_probabilities = class_probabilities_sum / np.sum(class_probabilities_sum)

# Print the overall probability for each class
for class_index, probability in enumerate(overall_class_probabilities):
    print(f"Class {class_index}: {probability*100:.2f}%")
```

Am observat că la SVC este posibil de menționat tipul kernelului de utilizat(dacă nu îl specificăm by default este RBF). Un kernel este o funcție matematică utilizată pentru a măsura similaritatea între două instanțe de date într-un spațiu dimensional superior.

- LINEAR kernel
Este cel mai simplu kernel și generează un hiperplan liniar ca frontieră de decizie.
Este potrivit pentru seturi de date care sunt liniar separabile, adică date care pot fi separate printr-o linie sau un plan în spațiul caracteristicilor.

- POLYNOMIAL kernel
Polynomial kernel este capabil să modeleze relații non-liniare între caracteristici.
Parametrul degree controlează complexitatea funcției polinomiale. O valoare mai mare a gradului poate duce la un model mai complex, dar poate fi, de asemenea, mai susceptibil la overfitting.

- RADIAL BASIS FUNCTION (RBF) kernel
RBF kernel este cel mai folosit kernel în SVM.
Este eficient în capturarea relațiilor complexe și non-liniare între caracteristici.
Parametrul gamma controlează cât de departe se extind influența fiecărui punct de date. Valori mai mari ale lui gamma conduc la modele mai complexe, dar pot duce și la overfitting.

- SIGMOID kernel
Sigmoid kernel este alt kernel non-liniar.
Este folosit mai puțin frecvent decât celelalte kernele și poate fi mai sensibil la variații ale parametrilor.
Parametrii gamma și coef0 controlează forma și scalarea funcției sigmoid.

De asemenea, kernele non-lineare ocupă multe resurse computaționale, ceea ce uneori este crucial pentru antrenare și se alege mai des un kernel linear, care generalizează mai bine datele

- În prima etapă a construcției arborelui de decizie, acesta începe cu un nod rădăcină care conține întregul set de date și de la el ese începe împărțirea pe subnoduri după un criteriu dat (gini, entropie)

- Criteriul gini/entropie este măsură de impuritate a nodului, adică există valori amestecate din fiecare clasă în nodul de decizie(nodurile intermediare care din nou trebuie divizat în baza criteriului)
    - Gini măsoară frecvența cu care orice element din setul de date va fi etichetat greșit atunci când este etichetat în mod aleatoriu.
    - Entropia este o măsură a informației care indică dezordinea caracteristicilor față de țintă.

- Arborele se oprește atunci când este dat max_depth la care trebuie să ajungă(pentru seturi de date enorme) ori atunci când ajunge la frunzele arborelui(nodurile pure care reprezintă clasa dată)

```python
from sklearn.tree import _tree
import numpy as np

def calculate_gini_for_node(tree, node_id, node_samples):
    if tree.children_left[node_id] == _tree.TREE_LEAF:
        return gini_index(node_samples)
    else:
        current_gini = gini_index(node_samples)
        # Accesarea caracteristicii și pragului pentru divizarea nodului curent
        feature_index = tree.feature[node_id]
        threshold = tree.threshold[node_id]
        # Divizarea setului de date în funcție de caracteristica și pragul nodului curent
        left_mask = X_train[:, feature_index] <= threshold
        right_mask = ~left_mask
        # Recursiv calculăm indicele Gini pentru copiii nodului curent
        left_gini = calculate_gini_for_node(tree, tree.children_left[node_id], node_samples & left_mask)
        right_gini = calculate_gini_for_node(tree, tree.children_right[node_id], node_samples & right_mask)
        # Returnăm indicele Gini pentru nodul curent
        return current_gini, left_gini, right_gini

# Parcurgerea recursivă a arborelui pentru a calcula indicele Gini pentru fiecare nod
root_node_id = 0
gini_for_nodes = calculate_gini_for_node(tree, root_node_id, np.ones(X_train.shape[0], dtype=bool))
print("Gini Index for each node:", gini_for_nodes)
```

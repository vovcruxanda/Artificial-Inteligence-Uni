# Cats vs Dogs Image Classification Model
> This notebook aims to make a Classification Model that will predict whether an image is an image of a cat or an image of a dog.

## Preprocessing the Data

### Splitting the Dataset into Training, Validation, and Test Dataset

> The module split folders allows us to split the folder into the desired output.

```python
!pip install split-folders
```

```python
import splitfolders
```

```python
input_folder='/kaggle/input/kaggle-cat-vs-dog-dataset/kagglecatsanddogs_3367a/PetImages'

# We will split the training folder into training and validation using the data generator
# So in this step we will only need to split it into training and test folders
split_ratio = (0.9, 0.0, 0.1)
```

```python
# This code will initiate the splits
splitfolders.ratio(
    input_folder,
    output='/kaggle/working/',
    seed=500,
    ratio=split_ratio,
    group_prefix=None,
)
```

### Creating the Dataset Generator
> The advantages of using Dataset Generator is that we can do rescaling or sometimes Data Augmentation easier.

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
# Preprocessing Parameters

img_h = img_w = 200
batch_size = 32
```

```python
# Train and Validation Generator

tvgen = ImageDataGenerator(
    rescale=1./255,
    data_format='channels_last',
    validation_split=0.3
) # We do not need to do any data augmentation because we got a lot of training data
```

```python
# Training Dataset

train_ds = tvgen.flow_from_directory(
    '/kaggle/working/train/',
    target_size=(img_h, img_w),
    batch_size=batch_size,
    class_mode='categorical',
    shuffle=True,
    seed=300,
    subset='training'
)

# Validation Dataset

val_ds = tvgen.flow_from_directory(
    '/kaggle/working/train/',
    target_size=(img_h, img_w),
    batch_size=105,
    class_mode='categorical',
    shuffle=True,
    seed=300,
    subset='validation'
)
```

## Creating the Image Classification Model

```python
import tensorflow as tf
```

```python
from tensorflow import keras
from tensorflow.keras import layers,regularizers
```

### Creating CNN Blocks
> In order to lessen redunduncy, the author will use subclassing

```python
# Hyperparameters

w_decay = 0.001
l_rate = 0.001
```

```python
# Creating CNN Block 1
# Conv2d --> Maxpool2D

class CNN1(layers.Layer):
    def __init__(self, node_num, k_size=(3,3), a_func='relu'):
        super(CNN1, self).__init__()
        self.conv = layers.Conv2D(node_num, k_size, activation=a_func, padding='same')
        self.max = layers.MaxPooling2D((2,2))
    
    def call(self, input_tensor, training=False):
        x = self.conv(input_tensor)
        x = self.max(x, training=training)
        return x
```

```python
# Creating CNN Block 2
# Conv2d --> BatchNorm --> activation relu

class CNN2(layers.Layer):
    def __init__(self, node_num, k_size=(3,3)):
        super(CNN2, self).__init__()
        self.conv = layers.Conv2D(node_num, 
                                k_size, 
                                padding='same',
                                kernel_regularizer=regularizers.l2(w_decay))
        self.bn = layers.BatchNormalization()
    
    def call(self, input_tensor, training=False):
        x = self.conv(input_tensor)
        x = self.bn(x, training=training)
        x = tf.nn.relu(x)
        return x
    
```

## Creating the Model

```python
model = keras.Sequential([
    # Input Layer
    keras.Input(shape=(img_h,img_w,3)),
    
    # CNN Layers
    CNN1(128),
    CNN2(64),
    CNN1(128),
    CNN2(64),
    
    # Dense Layers
    layers.Flatten(),
    layers.Dense(32, activation='relu'),
    layers.Dense(8, activation='relu'),
    layers.Dense(2, activation='softmax')
    ])
```

```python
model.summary()
```

```python
model.compile(
    optimizer=keras.optimizers.Adamax(l_rate),
    loss=keras.losses.CategoricalCrossentropy(),
    metrics=['accuracy']
)
```

### Creating Custom Callbacks

```python
# Stopping when target accuracy is met

class StopTrain(keras.callbacks.Callback):
    def on_epoch_end(self, epoch, logs=None):
        if logs.get('accuracy') > 0.9 and logs.get('val_accuracy') > 0.8:
            print('Target Accuracy is met!')
            self.model.stop_training = True
```

```python
# Listing all callbacks
cb_list = [StopTrain()]
```

```python
import math
```

```python
# Steps Per Epoch Calculation Function

tr_size = len(train_ds)
val_size = len(val_ds)

steps_per_epoch_cal = lambda x: int(math.ceil(1. * x/batch_size))
```

```python
# Calculating steps per epoch and validation steps

steps_per_epoch = steps_per_epoch_cal(tr_size)
val_steps = steps_per_epoch_cal(val_size)
```

```python
history = model.fit(
    train_ds,
    epochs=200,
    steps_per_epoch=steps_per_epoch,
    validation_data=val_ds,
    validation_steps=val_steps,
    verbose=2,
    callbacks=cb_list,
)
```

### Plotting Accurary and Loss Chart

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'val'], loc='upper left')
plt.show()
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'val'], loc='upper left')
plt.show()
```

### Evaluation of the Model

```python
# Evaluation on Validation Dataset

model.evaluate(val_ds)
```

```python
# Preprocessing of Test Dataset
testgen = ImageDataGenerator(
rescale=1./255
)

test_ds = testgen.flow_from_directory(
    '/kaggle/working/test/',
    target_size=(img_h, img_w),
    batch_size=batch_size,
    class_mode='categorical',
    subset='training'
)
```

```python
# Evaluation on Test Dataset

test_eval = model.evaluate(test_ds)

print(f"Loss on Test Dataset is {test_eval[0]:.2f} and Test Dataset Accuracy is {test_eval[1]:.2f}.")
```

## Using the Model to make Predictions

```python
from PIL import Image
from tensorflow.keras.preprocessing.image import img_to_array,load_img
import numpy as np
```

```python
predict_list = [
    '/kaggle/working/test/Cat/8732.jpg',
    '/kaggle/working/test/Dog/11073.jpg',
    '/kaggle/working/test/Cat/3196.jpg',
    '/kaggle/working/test/Dog/9106.jpg',
    '/kaggle/working/test/Cat/7216.jpg',
    '/kaggle/working/test/Dog/10485.jpg'
    ]
```

```python
train_ds.class_indices
```

```python
classes = ['Cat','Dog']
```

```python
def predictshow(x=0):
    path = predict_list[x]
    img = load_img(path, target_size=(img_h,img_w,3))
    X = img_to_array(img)
    X = np.expand_dims(X, axis=0)
    val = model.predict(X)
    print(val)
    img = Image.open(path)  
    plt.imshow(img)
    plt.title(label=str(classes[np.argmax(val)]))
    plt.show
```

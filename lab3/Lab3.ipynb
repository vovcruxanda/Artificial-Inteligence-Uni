**Min** — This represent the minium or least number in the dataset, I have represented it with min here.

Se pare că avem o problemă semnificativă cu valorile aberante. Pentru a rezolva acest lucru trebuie a fi sterse outliers-urile, dar pentru a le putea șterge trebuie mai întâi să le găsim. Deci metode de detectare a valorilor aberante sunt:
1. Tukey's IQR method
2. Standard deviation method
3. Z-score method
4. Modified z-score
5. Isolation Forest

Isolation Forest este un algoritm de învățare nesupravegheată care identifică anomalii prin izolarea valorilor aberante din date. Isolation Forest este, de asemenea, eficientă din punct de vedere computațional. Algoritmul se bazează pe principiul că anomaliile sunt observații care sunt puține și diferite, acest lucru ar trebui să le facă mai ușor de identificat. De aceea aleg Isolation Forest.

https://www.kaggle.com/code/marcinrutecki/outlier-detection-methods

**Cum lucrează Isolation Forest?**

Începe partiționarea datasetului prin alegerea aliatoare a valorii de partiționare între valoarea maximă și minimă. Partiționarea subseturilor ete iterativă și se așteaptă ca anomaliile să fie izolate în mai puține partiții, adică necesită mai puține diviziuni pentru a fi izolate. În consecință, anomaliile vor ajunge mai aproape de rădăcina arborilor, iar punctele de date normale vor fi izolate mai adânc în arbori.

Odată ce arborii sunt construiți, se calculează lungimea medie a traseului pentru fiecare punct de date pentru a ajunge la izolare. Punctele de date care au o lungime medie mai mică a traseului sunt considerate anomalii, deoarece necesită mai puține partiții pentru a fi izolate.

```python
from sklearn.ensemble import IsolationForest
df2 = df1.copy()
```

```python
# Construirea modelului
model=IsolationForest(n_estimators=150, max_samples='auto', contamination=float(0.1), max_features=1.0)
model.fit(df2)
```

```python
# Adaugarea 'scores' și 'anomaly' coloane la df
scores=model.decision_function(df2)
anomaly=model.predict(df2)

df2['scores']=scores
df2['anomaly']=anomaly

anomaly = df2.loc[df2['anomaly']==-1]
anomaly_index = list(anomaly.index)
print('Total number of outliers is:', len(anomaly))
```

```python
# ștergerea valorilor aberante
df2 = df2.drop(anomaly_index, axis = 0).reset_index(drop=True)
```

```python
#Verificarea setului de date
fig, axes = plt.subplots(nrows=1, ncols=2,figsize=(6,6))
fig.suptitle('Outliers\n', size = 25)

sns.boxplot(ax=axes[0], data=df2['Amount'], palette='Spectral').set_title("Amount")
sns.boxplot(ax=axes[1], data=df2['Frequency'], palette='Spectral').set_title("Frequency")

plt.tight_layout()
```

## Scalarea datelor

Se face acest pas pentru a ajusta valorile caracteristicilor astfel încât acestea să aibă o anumită scală sau interval specific. Motivele pentru care se face scalarea datelor includ:

1. **Uniformizarea unităților de măsură**: Scalarea datelor aduce caracteristicile în aceeași scala, ceea ce facilitează compararea și interpretarea lor.

2. **Stabilirea performanței algoritmilor**: Scalarea datelor poate îmbunătăți performanța algoritmilor de învățare automată, reducând variația în scala caracteristicilor, deoarece ei sunt sensibili la variațiile în această scală.

3. **Accelerarea convergenței algoritmilor**: Algoritmi care implică optimizarea, cum ar fi algoritmii de învățare a mașinilor sau de învățare profundă, pot convergi mai rapid către soluții în prezența datelor scalate. Acest lucru se datorează faptului că scalarea poate reduce dimensiunea gradienților și poate ajuta la evitarea problemelor legate de saturarea funcțiilor de activare.

4. **Reducerea sensibilității la valori extreme (outliers)**: Scalarea datelor poate reduce efectele pe care valorile extreme (outliers) le pot avea asupra performanței algoritmilor, deoarece aceste valori sunt adesea amplificate de caracteristicile cu scala mare.

# Determinarea numărului optim de clustere p/u K-Means

Selectarea numărului optim de clustere este cheia pentru aplicarea algoritmului de clustering la setul de date, cum ar fi gruparea k-means, care necesită ca utilizatorul să specifice numărul de clustere k care urmează să fie generate. Aceasta este o procedură oarecum arbitrară, unul dintre cele mai slabe aspecte ale efectuării analizei cluster.

Diferența majoră dintre metoda Elbow și a siluetei este că Elbow calculează doar distanța euclidiană, în timp ce silueta ia în considerare variabile precum varianța, asimetria, diferențele mari-scăzute etc.

## Metoda Elbow

Metoda constă în reprezentarea grafică a variației explicate în funcție de numărul de clustere și în alegerea cotului curbei ca număr de clustere care trebuie utilizat.

Metoda Elbow este o reprezentare grafică a găsirii lui "K" optim într-o grupare K-means. Aceasta funcționează prin găsirea WCSS (Within-Cluster Sum of Square), adică *suma distanței pătratice dintre punctele dintr-un cluster și centroidul clusterului*.

Cum lucrează?
Executăm gruparea k-means pe setul de date pentru o gamă de valori k și calculăm suma distanțelor pătratice (inerția) de la fiecare punct la centrul atribuit. Reprezentând grafic aceste distanțe în funcție de valorile k, punctul "cot" este cel în care creșterea clusterelor nu oferă o mai bună potrivire cu datele.

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

wcss = []

for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42, n_init=10)
    kmeans.fit(df3) #Inside the .fit() method, the KMeans algorithm calculates the centroids iteratively on data
    wcss.append(kmeans.inertia_)

plt.plot(range(1, 11), wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

## Metoda siluetei

Metoda siluetei medii măsoară calitatea unei grupări. Adică, determină cât de bine se află fiecare obiect în cadrul clusterului său. O lățime medie ridicată a siluetei indică o grupare bună.

Punctajul siluetei de 1 înseamnă că grupurile sunt foarte dense și bine separate. Scorul de 0 înseamnă că clusterele se suprapun. Scorul mai mic de 0 înseamnă că datele aparținând clusterelor pot fi greșite/incorecte. 

The silhouette score is a measure of how similar an object is to its own cluster (cohesion) and how different it is from other clusters (separation).

**Matematica din spatele acestui lucru implică câțiva pași:**

**Pasul 1:Calculul coeziunii.** Se măsoară distanța medie dintre un punct de date și toate celelalte puncte din același cluster. Această distanță medie (să o numim a) indică coeziunea în cadrul clusterului.

a(i) = (1 / (numărul de puncte din clusterul lui i - 1)) * suma(d(i, j)) [pentru toate punctele j din același cluster].

**Pasul 2: Calculul separării.** Se găsește cel mai apropiat cluster de clusterul punctului de date și se calculează distanța medie a punctului față de toate punctele din acest cluster vecin cel mai apropiat. Această distanță medie minimă va fi notată cu b.

b(i) = min (1 / (numărul de puncte din clusterul lui k)) * sum(d(i, j)) [pentru toate punctele j din clusterul vecin].

**Pasul 3: Calcularea scorului siluetei.** Acum, pentru marea dezvăluire, calculăm scorul siluetei (s) pentru fiecare punct de date luând diferența dintre separare și coeziune și normalizând-o cu valoarea maximă dintre cele două.

s(i) = (b(i) - a(i)) / max(a(i), b(i))

```python
from sklearn import datasets
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# Instantiate the KMeans for 3 clusters
km = KMeans(n_clusters=3, random_state=42, n_init=10)
# Fit the KMeans model
km.fit_predict(df3)
# Calculate Silhouette Score
score = silhouette_score(df3, km.labels_, metric='euclidean')
# Print the score
print('Silhouette Average Score: %.3f' % score)
```

```python
from sklearn import datasets
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# Instantiate the KMeans for 8 clusters
km = KMeans(n_clusters=8, random_state=42, n_init=10)
# Fit the KMeans model
km.fit_predict(df3)
# Calculate Silhouette Score
score = silhouette_score(df3, km.labels_, metric='euclidean')
# Print the score
print('Silhouette Average Score: %.3f' % score)
```

## K-Means

Clusterizarea K-Means implică atribuirea exemplelor la clustere în încercarea de a minimiza variația în cadrul fiecărui cluster. Este un algoritm bazat pe centroizi și cel mai simplu algoritm de învățare nesupravegheată. Algoritmul încearcă să minimizeze varianța punctelor de date din cadrul unui cluster. Este, de asemenea, modul în care majoritatea oamenilor sunt introduși în învățarea automată nesupravegheată.

K-means++ (parametrul de init implicit pentru K-Means în sklearn) este algoritmul care este utilizat pentru a depăși dezavantajul prezentat de algoritmul k-means. Scopul este de a răspândi centroidul inițial prin atribuirea primului centroid în mod aleatoriu, apoi prin selectarea restului centroizilor pe baza distanței pătratice maxime. Ideea este de a împinge centroizii cât mai departe posibil unul de celălalt.

Deși inițializarea în K-means++ este mai costisitoare din punct de vedere computațional decât algoritmul K-means standard, timpul de execuție pentru convergența spre optim este drastic redus pentru K-means++. Acest lucru se datorează faptului că este probabil ca centroizii aleși inițial să se afle deja în clustere diferite.

**AVANTAJELE grupării K-Means:**

1. **Simplu** - Este ușor de implementat

2. **Performanță ridicată** - Este rapidă și eficientă din punct de vedere al costurilor de calcul.

3. **Ușor de interpretat** - Produce clustere care pot fi ușor de interpretat și vizualizat.

4. **Potrivit pentru seturi mari de date** - În timp ce algoritmii de clusterizare sunt cunoscuți ca fiind relativ lenți, algoritmul k-means este relativ rapid, de aceea este eficient pentru seturi mari de date.

**DEZAVANTAJE ale clusterizării K-Means:**

1. **Presupune o densitate sferică** - acest lucru înseamnă că gruparea k-means nu funcționează la fel de bine în situațiile în care clusterele au în mod natural forme neregulate. Aceasta este o ipoteză relativ strictă.

2. **Sensibilă la scară** - dacă una dintre variabile este la o scară mult mai mare decât celelalte, acea variabilă va avea un efect foarte mare asupra distanței calculate. Acest lucru înseamnă că, în general, trebuie să redimensionăm datele înainte de a utiliza gruparea k-means.

3. **Dificil de încorporat variabilele categorice** - K-means este destinat să fie utilizat atunci când toate caracteristicile sunt numerice. Există modalități prin care vă puteți adapta datele pentru a fi adecvate dacă aveți unele caracteristici categorice, dar, în general, majoritatea caracteristicilor dvs. ar trebui să fie numerice.

4. **Sensibil la valorile aberante** - K-Means nu are o metodă de detectare a valorilor aberante. Centroizii pot fi târâți de valorile aberante sau valorile aberante ar putea primi propriul cluster în loc să fie ignorate.

5. **Se luptă cu datele cu dimensiuni mari** - Algoritmul depinde de distanța euclidiană, care este destul de proastă în cazul dimensiunilor mari. Dacă avem multe caracteristici potențiale, ar trebui să luăm în considerare aplicarea unor algoritmi de selecție a caracteristicilor sau de reducere a dimensionalității datelor înainte de a crea clustere.

6. **Algoritmul este randomizat** - ceea ce înseamnă că îl putem rula pe același set de date de mai multe ori și putem obține răspunsuri diferite. Aceasta este o problemă importantă pentru o mulțime de aplicații.

7. **k-means poate separa doar clusterele care sunt separabile mai mult sau mai puțin liniar** - Dacă clusterele dvs. se bazează pe distanța față de origine, k-means nu va putea să le identifice. Putem remedia această problemă trecând la coordonate polare.

Modalități de evitare a problemei sensibilității de inițializare în algoritmul k-means:

**Random** - se aleg n_clustere de observații (rânduri) aleatorii din date pentru centroizii inițiali.

**K-Means++** - o tehnică inteligentă de inițializare a centroizilor. Doar un singur centroid este inițializat aleatoriu, iar ceilalți centroizi sunt aleși astfel încât să fie foarte departe de centroidul/centralidele inițial/alegeri. Acest lucru are ca rezultat o convergență mai rapidă și o posibilitate mai mică de a avea un centroid prost inițializat. Inițializare implicită pentru sklearn.

K means++ oferă rezultate comparativ mai bune.

## Gruparea ierarhică

Gruparea ierarhică este o metodă de analiză a grupărilor care urmărește să construiască o ierarhie a grupărilor. Se utilizează pentru a grupa obiectele în clustere în funcție de cât de asemănătoare sunt între ele. În general, fuziunile și divizările sunt determinate într-o manieră greedy. Rezultatele grupării ierarhice sunt de obicei prezentate sub forma unei dendrograme.

Gruparea ierarhică este deosebit de utilă în situațiile în care sunt câteva observații care ne interesează în mod deosebit și dorim să putem identifica observațiile care sunt similare cu acestea.

Clusterizarea ierarhică poate fi:
- **Aglomerativă** - începe cu un element individual și apoi le grupează în clustere unice.
- **Divizivă** - începe cu un set de date complet și îl împarte în partiții.

Clusterizarea aglomerativă este cea mai bună pentru a găsi clustere mici.


**Avantajele grupării ierarhice:**

1. **Obțineți cele mai asemănătoare observații față de orice observații date** - algoritmul oferă informații detaliate cu privire la observațiile care sunt cele mai asemănătoare între ele. Acest nivel de detaliu nu este furnizat de mulți alți algoritmi.

2. **Nu este atât de sensibil la condițiile de inițializare** - nu este sensibil la condițiile de inițializare, cum ar fi semințele care sunt stabilite sau ordinea setului de date. În general, ar trebui să obțineți rezultate foarte asemănătoare și, în unele cazuri, același rezultat exact, dacă reluați analiza cu condiții de inițializare diferite.

3. **Poate fi adaptat pentru a încorpora variabile categorice** - poate fi adaptat pentru a sprijini situațiile în care aveți un amestec de variabile numerice și categorice relativ ușor. Pentru a face acest lucru, trebuie să vă asigurați că folosiți o metrică de distanță adecvată pentru tipuri de date mixte, cum ar fi distanța lui Grower.

4. **Mai puțin sensibilă la valorile aberante** - este puțin probabil ca prezența câtorva valori aberante să afecteze modul în care algoritmul se comportă cu celelalte puncte de date.

5. **Ipoteze mai puțin stricte cu privire la forma clusterelor** - algoritmii nu fac ipoteze la fel de stricte cu privire la forma clusterelor dumneavoastră. În funcție de metrica de distanță pe care o utilizați, unele forme de clustere pot fi detectate mai ușor decât altele, dar există mai multă flexibilitate.

**Dezavantajele grupării ierarhice:**

1. **Relativ lent** - este un algoritm foarte greu din punct de vedere matematic. În general, gruparea ierarhică necesită calcularea distanței pe perechi între toate observațiile dintr-un set de date, astfel încât numărul de calcule necesare crește rapid pe măsură ce crește dimensiunea setului de date.

2. **Trebuie să specificați numărul de clustere** - dar numărul de clustere poate fi modificat după ce partea principală a algoritmului a fost executată, astfel încât putem experimenta cu utilizarea unor numere diferite de clustere fără a fi nevoie să rulăm algoritmul de la zero.

3. **Sensibil la scară** - este posibil să trebuiască să redimensionăm datele înainte de a rula clusterizarea. Nivelul exact de sensibilitate va varia în funcție de metrica de distanță pe care o folosim pentru a calcula distanța dintre puncte.

4. **Foarte orientat către euristică** - acest lucru duce la o mulțime de intervenții manuale în proces și, în consecință, sunt necesare cunoștințe specifice aplicației/domeniului pentru a analiza dacă rezultatul are sens sau nu.

5. **Posibil dificil de vizualizat** - Dacă numărul de eșantioane de date crește, atunci analizarea vizuală a dendrogramei și luarea deciziilor devine imposibilă.

**linkage{'ward', 'complete', 'average', 'single'}, default='ward'**
Este criteriul de corelare care trebuie utilizat. Criteriul de corelare determină distanța care trebuie utilizată între seturile de observații. Algoritmul va fuziona perechile de clustere care minimizează acest criteriu.

**ward** minimizează varianța dintre clusterele care sunt fuzionate. is the most effective method for noisy data.

**average** (medie) utilizează media distanțelor fiecărei observații din cele două seturi.

**complete** utilizează distanțele maxime dintre toate observațiile din cele două seturi.

**average and complete** linkage perform well on cleanly separated globular clusters, but have mixed results otherwise.

**single** utilizează distanțele minime dintre toate observațiile din cele două seturi. is fast, and can perform well on non-globular data, but it performs poorly in the presence of noise.

## Algoritmul de clusterizare DBSCAN

DBSCAN reprezintă gruparea spațială bazată pe densitate a aplicațiilor cu zgomot. Este un algoritm de clusterizare bazat pe densitate. Este capabil să găsească clustere de formă neregulată. Separă regiunile prin zone de densitate scăzută, astfel încât poate detecta foarte bine și valorile aberante. Acest algoritm este mai bun decât k-means atunci când vine vorba de lucrul cu date cu forme ciudate.


**Avantajele DBSCAN:**

1. **Gestionează clusterele de formă și dimensiuni neregulate** - DBSCAN este unul dintre algoritmii care face cele mai puține presupuneri cu privire la forma clusterelor.

2. **Robust la valorile aberante** - algoritmul este capabil să detecteze valorile aberante și să le excludă complet din clustere.

3. **Nu necesită specificarea numărului de clustere** - un alt avantaj al DBSCAN este că nu necesită ca utilizatorul să specifice numărul de clustere. DBSCAN poate detecta automat numărul de clustere care există în date. Acest lucru este excelent pentru cazurile în care nu aveți prea multă intuiție cu privire la numărul de clustere care ar trebui să existe.

4. **Mai puțin sensibil la condițiile de inițializare** - DBSCAN este mai puțin sensibil la condițiile de inițializare, cum ar fi ordinea observațiilor din setul de date și sămânța utilizată, decât alți algoritmi de clusterizare.

5. **Relativ rapid** - DBSCAN este, în general, mai lent decât clusterizarea K-Means, dar mai rapid decât clusterizarea ierarhică și clusterizarea spectrală.

**Dezavantajele DBSCAN:**

1. **Este dificil să încorporați caracteristici categorice** - ar trebui să utilizați DBSCAN în cazurile în care majoritatea caracteristicilor sunt numerice.

2. **Necesită o scădere a densității pentru a detecta granițele dintre clustere** - trebuie să existe o scădere a densității punctelor de date între clustere pentru ca algoritmul să poată detecta granițele dintre clustere. În cazul în care există mai multe clustere care se suprapun fără o scădere a densității datelor între ele, acestea pot fi grupate într-un singur cluster.

3. **Se luptă cu clusterele de densitate variabilă** - DBSCAN determină unde încep și unde se opresc clusterele, analizând locurile în care densitatea punctelor de date scade sub un anumit prag. Poate fi dificil să se găsească un prag care să capteze toate punctele din clusterul mai puțin dens fără a exclude prea multe valori aberante străine din clusterul mai dens.

4. **Sensibil la scară** - DBSCAN este sensibil la scara variabilelor. Aceasta înseamnă că trebuie să redimensionăm variabilele dacă acestea se află la scări foarte diferite.

5. **Se luptă cu datele cu dimensiuni mari** - performanța DBSCAN tinde să se degradeze în situațiile în care există multe caracteristici. Ar trebui să folosim tehnici de reducere a dimensionalității sau de selecție a caracteristicilor pentru a reduce numărul de caracteristici dacă avem un set de date cu dimensiuni mari.

6. **Este necesar să setați parametrii pentru Epsilon și MinPts.** Acest lucru ar putea fi dificil.

**Alegerea parametrilor inițiali potriviți**
DBSCAN necesită doi parametri - alegerea numerelor corecte este esențială pentru acest algoritm:

**Epsilon (ε)** *este raza cercului care trebuie creat în jurul fiecărui punct de date pentru a verifica densitatea.*
Această tehnică calculează distanța medie dintre fiecare punct și cei mai apropiați k vecini ai săi, unde k = valoarea MinPts selectată. Distanțele medii k sunt apoi reprezentate în ordine crescătoare pe un grafic al distanței k. Valoarea lui epsilon poate fi stabilită pe baza graficului de distanțe K. Punctul de curbură maximă (cotul) din acest grafic ne indică valoarea lui epsilon (adică locul în care graficul are cea mai mare pantă).

**minPoints (MinPts)** *este numărul minim de puncte de date necesare în interiorul cercului respectiv pentru ca acel punct de date să fie clasificat ca punct central.*
Nu există o modalitate automată de determinare a valorii MinPts pentru DBSCAN. Nu are sens să se ia minPoints ca fiind 1, deoarece astfel fiecare punct va constitui un cluster separat. Prin urmare, trebuie să fie de cel puțin 3. În general, este de două ori dimensiunile sau dimensiunile +1. Cunoașterea domeniului decide, de asemenea, valoarea sa.

Dacă setul de date este mai zgomotos, alegeți o valoare mai mare pentru MinPts
Pentru date bidimensionale, utilizați valoarea implicită a DBSCAN de MinPts = 4 (Ester et al., 1996).
Valoarea lui minPoints trebuie să fie cu cel puțin una mai mare decât numărul de dimensiuni al setului de date:

MinPts >= Dimensions * 2

SAU

MinPts >= Dimensions + 1

IMPORTANT:

Uneori este dificil să se găsească parametrii Epsilon și MinPts adecvați pentru setul de date. Acestea sunt câteva sfaturi:

- Utilizați un MinPts mai mare pentru seturi de date mari și zgomotoase.

- Dacă obțineți clustere prea mari: reduceți epsilon.

- Dacă obțineți prea mult zgomot: creșteți epsilon).

- O clusterizare bună necesită iterații.

# Expectation Maximization cu GMM(Gaussian Mixture Models)

Modelele mixte gaussiene sunt un model probabilistic pentru reprezentarea subpopulațiilor distribuite normal în cadrul unei populații globale. În general, modelele mixte nu necesită cunoașterea subpopulației din care face parte un punct de date, permițând modelului să învețe subpopulațiile în mod automat.

În cel mai simplu caz, GMM-urile pot fi utilizate pentru a găsi clustere în același mod ca și k-means, dar, deoarece GMM conține un model probabilistic, este posibil să se găsească, de asemenea, atribuiri probabilistice de clustere.




**Avantajele modelelor de amestecuri gaussiene (GMM):**

1. **Estimări probabilistice ale apartenenței la fiecare cluster** - modelele oferă estimări ale probabilității ca fiecare punct de date să aparțină fiecărui cluster. Aceste estimări de probabilitate pot fi foarte utile atunci când se examinează puncte de date ambigue care se află la granița a două clustere.

2. **Nu presupune clustere sferice** - modelul nu presupune că toate clusterele sunt sfere de formă uniformă. În schimb, modelele de amestec gaussian pot fi utilizate pentru a acomoda clusterele de forme diferite.

3. **Gestionează clustere de dimensiuni diferite**

4. **Mai puțin sensibil la scară** - nu este nevoie de redimensionat variabilele înainte de a le utiliza pentru clusterizare.

5. **Găzduiește apartenența mixtă** - în kmeans, un punct aparține unui singur cluster, în timp ce în GMM un punct aparține fiecărui cluster într-o măsură diferită. Gradul se bazează pe probabilitatea ca punctul să fie generat din distribuția normală (multivariată) a fiecărui cluster, cu centrul clusterului ca medie a distribuției și covarianța clusterului ca și covarianță a acesteia. În funcție de sarcină, apartenența mixtă poate fi mai potrivită (de exemplu, articolele de știri pot aparține mai multor clustere de subiecte) sau nu (de exemplu, organismele pot aparține unei singure specii).

**Dezavantajele modelului de amestec gaussian (GMM):**

1. **Dificil de încorporat caracteristici categorice** - modelele funcționează în baza ipotezei că toate caracteristicile sunt distribuite în mod normal, astfel încât nu sunt ușor de adaptat la datele categorice.

2. **Presupune o distribuție normală pentru caracteristici** - acest lucru înseamnă că ar trebui să vă acordați ceva timp pentru a analiza distribuțiile caracteristicilor înainte de a ajunge la acest algoritm de clusterizare.

3. **Face unele presupuneri cu privire la forma clusterelor** - acest lucru înseamnă că modelele de amestec gaussian nu vor funcționa la fel de bine în cazurile în care clusterele au o formă foarte neregulată.

4. **Are nevoie de suficiente date pentru fiecare cluster** - trebuie să vă asigurați că aveți suficiente puncte de date în fiecare cluster pentru a estima în mod adecvat covarianța. Cantitatea de date necesară nu este uriașă, dar este mai mare decât cea a algoritmilor simpli care nu estimează o matrice de covarianță.

5. **Necesitatea de a specifica numărul de clustere** - deoarece modelele de amestec gaussian funcționează în baza ipotezei că caracteristicile dvs. sunt distribuite normal, acestea pot fi dezorientate în cazurile în care există multe valori aberante în date. Unele implementări ale modelelor de amestec gaussian permit separarea valorilor aberante într-un cluster separat.

6. **Sensibil la condițiile de inițializare** - puteți obține rezultate diferite dacă rulați algoritmul de mai multe ori.

7. **Lent** - acest lucru este valabil mai ales atunci când există multe caracteristici în setul dvs. de date.

Modelul presupune că datele sunt generate dintr-o combinație de mai multe distribuții gaussiene (sau normale). Fiecare distribuție Gaussiană din combinație reprezintă un cluster în date.

**1.** GMM presupune că punctele de date din fiecare cluster sunt centrate în jurul unei valori medii și au o anumită dispersie sau varianță. Adică distribuția generală a datelor este o combinație de mai multe distribuții gaussiene

**2.** Funcția de densitate probabilistică este o sumă ponderată a funcțiilor de densitate probabilistică ale distribuțiilor gaussiane individuale.



P(x) este funcția de densitate probabilistică a datelor.

K este numărul de clustere.

Pi este ponderea componentei gaussiene i, reprezentând probabilitatea de a selecta acea componentă.

N(x|μi, Σi) este distribuția gaussiană cu medie μi și matrice de covarianță Σi.

**3.** Estimarea parametrilor precum media, covarianța și ponderilor se face prin Expectation Maximization. Algoritmul EM estimează iterativ parametrii prin maximizarea verosimilității datelor observate.

**4.** Odată ce parametrii sunt estimati, GMM poate fi folosit pentru a atribui etichete de cluster punctelor de date.
https://www.kaggle.com/code/vipulgandhi/gaussian-mixture-models-clustering-explained

**Expectation**

Prima etapă constă în calcularea așteptării atribuțiilor componentelor **Ck** pentru fiecare punct de date xi∈X, având în vedere parametrii modelului πk μk și σk.

**Maximizare**

A doua etapă este cunoscută sub denumirea de etapă de maximizare sau M care constă în maximizarea așteptărilor calculate în etapa E în ceea ce privește parametrii modelului. Această etapă constă în actualizarea valorilor πk, μk și σk.

Întregul proces iterativ se repetă până când algoritmul converge, oferind o estimare de maximă verosimilitate. În mod intuitiv, algoritmul funcționează deoarece cunoașterea atribuirii componentelor Ck pentru fiecare xi face ca rezolvarea pentru πk μkși σk este ușor, în timp ce cunoașterea lui πk μk σk permite să se deducă p(Ck|xi) ușor. Etapa de așteptare corespunde celui de-al doilea caz, în timp ce etapa de maximizare corespunde primului caz. Astfel, prin alternanța între valorile presupuse fixe sau cunoscute, se pot calcula în mod eficient estimările de maximă verosimilitate ale valorilor nefixate.

```python
from sklearn.mixture import GaussianMixture
from sklearn.metrics import silhouette_score

# Initialize Gaussian Mixture Model
gmm = GaussianMixture(n_components=6, random_state=42)

# Fit the model to the data
gmm.fit(df3)
```

```python
# Predict the cluster assignments for each data point
y_gmm = gmm.predict(df3)
```

```python
# Evaluate the clustering performance using silhouette score
silhouette_score_gmm = silhouette_score(df3, y_gmm)
print("Silhouette Score (Gaussian Mixture Model):", silhouette_score_gmm)
```

```python
# Add the cluster labels to the dataframe
df_gmm = df2.copy()
df_gmm['Cluster'] = y_gmm

# Print the number of items in each cluster
cluster_counts = df_gmm['Cluster'].value_counts()
print("Cluster Sizes:")
print(cluster_counts)
```

```python
# Calculate the mean for each column
mean_values = gmm.means_

# Print the mean values
print("Mean values for each column:")
print(mean_values)
```

Aceste mean-uri reprezintă punctele în spațiul caracteristicilor care sunt considerate a fi "centrul" fiecărui cluster. Mi s-a părut straniu valorile negative, dar din ceea ce am găsit acest lucru reprezintă o normalitate. Atunci când se efectuează clusterizarea folosind un GMM, mean-urile negative pentru unele dimensiuni pot indica că centrul unor clustere se află în partea negativă a acelei dimensiuni.

Covarianța este o măsură a relației între două variabile aleatoare. Aceasta indică modul în care variază două variabile împreună. Dacă o creștere a uneia dintre variabile este asociată cu o creștere a celeilalte variabile, covarianța este pozitivă. În schimb, dacă o creștere a uneia dintre variabile este asociată cu o scădere a celeilalte variabile, covarianța este negativă.

```python
# Calculate the standard deviation for each column
covariances = gmm.covariances_

# Print the standard deviation values
print("Standard deviation for each column:")
print(covariances)
```
